## Best defensive, midfield and offensive players: 

Highest  defensive, midfield and offensive score of each team. The average of each is based on the average of all the players in their respective position, the dataset already provides this based on the data of the FIFA videogame.

```python
score_columns = df.drop(columns=df.columns[3:19])
score_columns = score_columns.drop(columns=df.columns[0])
score_columns  
```

```python
countries_score = {}
# temp_df = pd.DataFrame()
for index, row in score_columns.iterrows():
  country = row['home_team']
  if country not in countries_score:
    countries_score[country] = score_columns.iloc[index,2:5].values.astype('float32')
   
    country = row['away_team']
  if country not in countries_score:
    countries_score[country] = score_columns.iloc[index,5:8].values.astype('float32')
```

```python
countries_score#{'country':[mean_defensive_score, mean_midfield_score, mean_offense_score]}
#we pick the most recent scores
```

```python
team_defmidof_score = pd.DataFrame.from_dict(countries_score, orient='index',columns=['defensive_score', 'midfield_score', 'offensive_score'])
team_defmidof_score
```

```python
defns_score = team_defmidof_score['offensive_score'].iloc[0]
print(defns_score)
best_off_country = {'Korea Republic':team_defmidof_score['offensive_score'].iloc[0]}
print(type(best_off_country['Korea Republic']))
labelx = team_defmidof_score.axes[0].tolist()
print(labelx)
```

```python
best_def_country = ['Korea Republic', team_defmidof_score['defensive_score'].iloc[0]]#-->initialize with the first one
best_mid_country = ['Korea Republic', team_defmidof_score['defensive_score'].iloc[0]]#-->initialize with the first one
best_off_country = ['Korea Republic', team_defmidof_score['defensive_score'].iloc[0]]#-->initialize with the first one
for index, row in team_defmidof_score.iterrows():
  
  defns_score = row['defensive_score']
  mid_score = row['midfield_score']
  off_score  = row['offensive_score']

  if defns_score >= best_def_country[1]:
    best_def_country[0] = row.name
    best_def_country[1] = defns_score

  if mid_score >= best_mid_country[1]:
    best_mid_country[0] = row.name
    best_mid_country[1] = mid_score

  if off_score >= best_off_country[1]:
    best_off_country[0] = row.name
    best_off_country[1] = off_score
```

##All goals since 1993

```python
goals_columns = df.drop(columns=df.columns[3:9])
goals_columns = goals_columns.drop(columns=goals_columns.columns[5::])
goals_columns = goals_columns.drop(columns=goals_columns.columns[0])
goals_columns  
```

```python
#->initialize the countries in the dictionary
countries_total_goals = {}
for index, row in goals_columns.iterrows():
  country = row['home_team']
  if country not in countries_total_goals:
    countries_total_goals[country] = 0

    country = row['away_team']
  if country not in countries_total_goals:
    countries_total_goals[country] = 0
  
countries_total_goals
```

```python
#Add the goals inside the dict
for index, row in goals_columns.iterrows():
  country = row['home_team']
  if country in countries_total_goals:
    countries_total_goals[country] += row['home_team_score']
   
    country = row['away_team']
  if country in countries_total_goals:
    countries_total_goals[country] += row['away_team_score']

countries_total_goals
```

```python
goals_per_country = pd.DataFrame()
goals_per_country = goals_per_country.from_dict(data=countries_total_goals,orient='index',columns=['goals'])
goals_per_country = goals_per_country.rename_axis("country").reset_index()
goals_per_country
```

```python
fig = plt.figure(figsize=(30, 10))
sns.scatterplot(data=goals_per_country,x='country',y='goals')
plt.xticks(rotation=90);
plt.title('All historic goals since 1993');
```

## Goals since early 2021

```python
goals_columns_bydate = df.drop(columns=df.columns[3:9])
goals_columns_bydate = goals_columns_bydate.drop(columns=goals_columns_bydate.columns[5::])
goals_columns_bydate 
```

```python
import datetime
goals_columns_bydate['date'] =  pd.to_datetime(goals_columns_bydate['date'])

goals_columns_early2021 = pd.DataFrame()
limit_datetime = datetime.datetime(2021,1,1)
for index, row in goals_columns_bydate.iterrows():
  if row['date']<=limit_datetime:
    break
  else:
    goals_columns_early2021 = goals_columns_early2021.append(row)

goals_columns_early2021 #--> we have 1647 international matches since the early 2021
```

```python
#->initialize the countries in the dictionary
countries_goals_earlys2021 = {}
for index, row in goals_columns_early2021.iterrows():
  country = row['home_team']
  if country not in countries_goals_earlys2021:
    countries_goals_earlys2021[country] = 0

    country = row['away_team']
  if country not in countries_goals_earlys2021:
    countries_goals_earlys2021[country] = 0
  
#Add the goals inside the dict
for index, row in goals_columns_early2021.iterrows():
  country = row['home_team']
  if country in countries_goals_earlys2021:
    countries_goals_earlys2021[country] += row['home_team_score']
   
    country = row['away_team']
  if country in countries_goals_earlys2021:
    countries_goals_earlys2021[country] += row['away_team_score']

countries_goals_earlys2021
```

```python
goals_per_country_earlys2021 = pd.DataFrame()
goals_per_country_earlys2021 = goals_per_country_earlys2021.from_dict(data=countries_goals_earlys2021,orient='index',columns=['goals'])
goals_per_country_earlys2021 = goals_per_country_earlys2021.rename_axis("country").reset_index()
goals_per_country_earlys2021
```

```python
fig = plt.figure(figsize=(50, 10))
sns.scatterplot(data=goals_per_country_earlys2021,x='country',y='goals')
plt.xticks(rotation=90);
plt.title('All historic goals since earlys 2021');
```
