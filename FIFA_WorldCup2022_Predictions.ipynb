```python
import pandas as pd
```

```python
url = 'https://drive.google.com/file/d/1f-skbu-y4GAte9aUWWLBwYFQ78ECiS3n/view?usp=sharing'
url='https://drive.google.com/uc?id=' + url.split('/')[-2]
df = pd.read_csv(url)
```

```python
df.head()
```

```python
df = df.sort_values(by='date',ascending=False)
df = df.reset_index(drop=True)
df
```

```python
score_columns = df.drop(columns=df.columns[3:19])
score_columns = score_columns.drop(columns=df.columns[0])
score_columns  
```

```python
score_columns.iloc[0,5:8].values
```

```python
countries_score = {}
temp_df = pd.DataFrame()
for index, row in score_columns.iterrows():
  country = row['home_team']
  if country not in countries_score:
    countries_score[country] = score_columns.iloc[index,2:5].values.astype('float32')
   
    country = row['away_team']
  if country not in countries_score:
    countries_score[country] = score_columns.iloc[index,5:8].values.astype('float32')
```

```python
countries_score#{'country':[mean_defensive_score, mean_midfield_score, mean_offense_score]}
#we pick the most recent scores
```

```python
team_defmidof_score = pd.DataFrame.from_dict(countries_score, orient='index',columns=['defensive_score', 'midfield_score', 'offensive_score'])
team_defmidof_score
```

```python
defns_score = team_defmidof_score['offensive_score'].iloc[0]
print(defns_score)
best_off_country = {'Korea Republic':team_defmidof_score['offensive_score'].iloc[0]}
print(type(best_off_country['Korea Republic']))
labelx = team_defmidof_score.axes[0].tolist()
print(labelx)
```

```python
best_def_country = {'Korea Republic':team_defmidof_score['defensive_score'].iloc[0]}

best_def_country = ['Korea Republic', team_defmidof_score['defensive_score'].iloc[0]]#-->initialize with the first one
best_mid_country = ['Korea Republic', team_defmidof_score['defensive_score'].iloc[0]]#-->initialize with the first one
best_off_country = ['Korea Republic', team_defmidof_score['defensive_score'].iloc[0]]#-->initialize with the first one
for index, row in team_defmidof_score.iterrows():
  
  defns_score = row['defensive_score']
  mid_score = row['midfield_score']
  off_score  = row['offensive_score']

  if defns_score >= best_def_country[1]:
    best_def_country[0] = row.name
    best_def_country[1] = defns_score

  if mid_score >= best_mid_country[1]:
    best_mid_country[0] = row.name
    best_mid_country[1] = mid_score

  if off_score >= best_off_country[1]:
    best_off_country[0] = row.name
    best_off_country[1] = off_score
```
